# Plot using numpy

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
x = np.arange(5, 10) 
y = np.arange(12, 17)
```

```python
x
```

```python
y
```

```python
plt.plot(x,y) 
plt.title("Matplotlib PLot NumPy Array") 
plt.xlabel("x axis") 
plt.ylabel("y axis") 
plt.show()
```

```python
plt.plot(x,y, linestyle='--',color='green') 
plt.title("Matplotlib Line PLot NumPy Array") 
plt.xlabel("x axis") 
plt.ylabel("y axis") 
plt.show()
```

```python
plt.scatter(x,y,color='green') 
plt.title("Matplotlib Scatter PLot NumPy Array") 
plt.xlabel("x axis") 
plt.ylabel("y axis") 
plt.show()
```
